Já que $0=e^x - 5sen x$ => $sen x = \frac{e^x}{5}$, não existe uma regularidade para definir um intervalo como as raízes de $sen x$ ($x^* = n*\pi | n \in Z$)

Mas, sempre existe uma raíz nos intervalos $[\lfloor n\pi \rfloor, \lceil n\pi \rceil]$, para $n \in Z^-$, já que $\lim_{x \to -\infty} e^x = 0$

Exemplo: \[-4,-3\]

pode-se fazer o seguinte procedimento: (Método Gráfico).

#### 5) Isole e calcule uma raiz de cada função, utilizando os métodos indicados na tabela. <a id='5'></a>
Use precisão $\epsilon$ = $10^{-2}$. Compare o número de iterações de cada método. Pode-se
afirmar que existe um método definitivamente melhor que todos os demais?

```python
# Primeira
def f(x) :
    return math.e**np.cos(x)+x**3-3
def fp(x) :
    return (3*x**2)-(np.sin(x)*(math.e**np.cos(x)))

x = np.linspace(-1,2,1000)
plot.plot(x,f(x))

a = 1
b = 1.5

r1 = [bissection.findRoot(f,a,b,1e-2),strings.findRoot(f,a,b,1e-2),newton.findRoot(f,fp,a,b,1e-2)]

print(r1)
```

```python
# Segunda
def f(x) :
    return 0.1*x**3-math.e**(2*x)+2
def fp(x) :
    return 0.3*x**2-(2*math.e**(2*x))

x = np.linspace(-3,1,1000)
plot.plot(x,f(x))

a = -0.5
b = 1

r2 = [bissection.findRoot(f,a,b,1e-2),strings.findRoot(f,a,b,1e-2),newton.findRoot(f,fp,a,b,1e-2)]

print(r2)
```

```python
# Terceira
def f(x) :
    return x**3-5*x**2+x+3
def fp(x) :
    return 3*x**2-10*x+1

x = np.linspace(-2,5,1000)
plot.plot(x,f(x))

a = 4
b = 5

r3 = [bissection.findRoot(f,a,b,1e-2),strings.findRoot(f,a,b,1e-2),newton.findRoot(f,fp,a,b,1e-2)]

print(r3)
```

```python
display(Markdown("""
|           Função           | Bisseção |    | Cordas   |    | Newton   |    |
|----------------------------|----------|----|----------|----|----------|----|
|                            | raiz     | n  | raiz     | n  | raiz     | n  |
| $$y = e^{{cos(x)}}+x^3-3$$ | {:.6f}   | {} | {:.6f}   | {} | {:.6f}   | {} |
| $$y = 0.1x^3-e^{{2x}}+2$$  | {:.6f}   | {} | {:.6f}   | {} | {:.6f}   | {} |
| $$y = x^3-5x^2+x+3$$       | {:.6f}   | {} | {:.6f}   | {} | {:.6f}   | {} |
""".format(r1[0][0],r1[0][1],r1[1][0],r1[1][1],r1[2][0],r1[2][1],r2[0][0],r2[0][1],r2[1][0],r2[1][1],r2[2][0],r2[2][1],r3[0][0],r3[0][1],r3[1][0],r3[1][1],r3[2][0],r3[2][1])))
```

Apesar de em algumas instâncias algum método poder se sobressair aos outros, é difícil afirmar que
um método é melhor que outro em todas as situações. Podem acontecer situações como o método de Newton
ser incalculável, no caso da derivada da função ser 0 na divisão, e de o método da bisseção achar muito
rapidamente um ponto caso ele esteja exatamente no meio do intervalo. Além disso, computacionalmente,
várias iterações de um método podem ser computadas mais rápido que uma única iteração de um método mais
computacionalmente caro.

Devido a isto, não se pode afirmar com convicção que um método é sempre melhor que outro método,
mas pode-se afirmar que um método é melhor que outro dadas as devidas restrições e intenções.

```python
# c
newton.printRoot(f,fp,-1,-0.25,1e-2)
newton.printRoot(f,fp,0.5,1.5,1e-2)
newton.printRoot(f,fp,4,5,1e-2)
```

```python
# d
a = 4
b = 5

md = """
|  n  |  x  | f(x) |
|-----|-----|------|
"""

xm = (a+b)/2
if fp(xm)*newton.derivApprox(fp,xm)>0 :
    x = b
else :
    x = a
x = newton.newton(f,fp,x)
n = 0
md += "| {} | {:.6f} | {:.6f} |\n".format(n,x,f(x))
while abs(f(x)) > 1e-6 :
    x = newton.newton(f,fp,x)
    n+=1
    md += "| {} | {:.12f} | {:.12f} |\n".format(n,x,f(x))

display(Markdown(md))
```

```python
# e

md = """
|  n  |  a  |  b  |  x  | f(x) |
|-----|-----|-----|-----|------|
"""

x = (a+b)/2
n = 0
md += "| {} | {:.12f} | {:.12f} | {:.12f} | {:.12f} |\n".format(n,a,b,x,f(x))
while abs(f(x))>1e-6 :
    fx = f(x)
    if fx*f(a)<0 : b = x
    else : a = x
    x = (a+b)/2
    n+=1
    md += "| {} | {:.12f} | {:.12f} | {:.12f} | {:.12f} |\n".format(n,a,b,x,f(x))

display(Markdown(md))
```

Convergem de forma diferente. O método de newton sempre se aproxima pelo mesmo "lado" da raiz, enquanto o método da bisseção pode acabar indo para um lado oposto.

#### 7) O preço à vista (PV) de uma mercadoria é R\\$ 312.000,00, mas pode ser financiado com uma entrada (E) de R\\$ 91.051,90 e 12 (P) prestações mensais (PM) de R\\$ 26.000,00. Calcule a taxa de juros (j) sabendo que <a id='7'></a>

$$\frac{1-(1+j)^{-P}}j=\frac{PV-E}{PM}$$

```python
PV = 312000
E = 91051.9
P = 12
PM = 26000

def f(j) :
    return (1-(1+j)**-P)/j - (PV-E)/PM

x = np.linspace(1e-3,0.5,1000)
plot.plot(x,f(x))

bissection.printRoot(f,1e-3,0.5,1e-6)
```

#### 8) Uma bola é arremessada para cima com velocidade $v_o = 30 m/s$, a partir de uma altura $x_o = 5 m$, em um local onde a aceleração da gravidade é $g = -9,81 m/s^2$,  sabendo que: <a id='8'></a>

Qual será o tempo gasto para a bola tocar no solo, desconsiderando o atrito com ar.

#### 9) Determinar o comprimento (L) de um cabo suspenso em dois pontos do mesmo nível e distantes (2x)400m, com flexa ($f$) de 100m, sabendo que <a id='9'></a>

<center>$L = 2a senh(\frac{x}{a})$, sendo $a$ a raíz da equação $a(cosh(\frac{x}{a})-1)-f=0$</center>

```python
distancia = 800 #distância
flexa = 100

def F(a) :
    return a*(np.cosh(distancia/a)-1)-flexa
def Fprime(a) :
    return -(distancia/a)*np.sinh(distancia/a)+np.cosh(distancia/a)-1

xarr = np.linspace(1000,5000,1000)
plot.plot(xarr,F(xarr))

a, c = newton.findRoot(F,Fprime,3000,3500,1e-6)
print(f"Root found at {a} ({F(a)}) with {c} iterations")

print(f"O comprimento (L) é {2*a*np.sinh(distancia/a)}")
```

Importar as biblioteas

```python
%matplotlib inline
%reload_ext autoreload
%autoreload 2

import plot
import numpy as np
```

Definir a função

```python
f = lambda x : x**3+2*x**2+4

leftlimit = -5
rightlimit = 5

xarr = np.linspace(leftlimit,rightlimit,1000)
yarr = f(xarr)
plot.plot(xarr,yarr)
```

Definir limites

```python
a = -3
b = 1
e = 1e-7 #function value considered as 0
plot.plotrange(xarr,yarr,a,b)
```

Algoritmo de bisseção

```python
x = (a+b)/2
n = 0
r = [[x,n]]
while abs(f(x))>e :
    fx = f(x)
    if fx==0 : break
    #print(a,b,x,f(a),f(b),fx)
    if fx*f(a)<0 : b = x
    elif fx*f(b)<0 : a = x
    else :
        raise Exception("Sem Raiz no intervalo")
    x = (a+b)/2
    n += 1
    r.append([x,n])
    
print("A raiz aproximada é: {} ({:.16f})".format(x,f(x)))
plot.plotdot(xarr, yarr, [(x,0)])

for x in r :
    print(x[1], x[0], f(x[0]))
```
